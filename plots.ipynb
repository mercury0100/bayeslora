```python
import matplotlib.pyplot as plt
import numpy as np

log_alpha = np.array([2, 3, 4, 5])

datasets = {
    "QQP": {
        "ada": [0.9035, 0.9083, 0.9083, 0.9116],
        "bayes": [0.9052, 0.9091, 0.9112, 0.9119],
        "ada_err": [0.0004, 0.0005, 0.0005, 0.0005],
        "bayes_err": [0.0006, 0.0008, 0.0007, 0.0005]
    },
    "QNLI": {
        "ada": [0.9408, 0.9425, 0.9429, 0.9433],
        "bayes": [0.9406, 0.9440, 0.9440, 0.9460],
        "ada_err": [0.0004, 0.0004, 0.0004, 0.0004],
        "bayes_err": [0.0012, 0.0011, 0.0011, 0.0014]
    },
    "SST2": {
        "ada": [0.9530, 0.9530, 0.9541, 0.9541],
        "bayes": [0.9570, 0.9590, 0.9570, 0.9560],
        "ada_err": [0.0003, 0.0003, 0.0003, 0.0003],
        "bayes_err": [0.0010, 0.0015, 0.0013, 0.0010]
    }
}

plt.rcParams.update({
    "font.size": 14,
    "axes.labelsize": 14,
    "axes.titlesize": 14,
    "legend.fontsize": 13,
    "xtick.labelsize": 13,
    "ytick.labelsize": 13
})

#  Only this changed: narrower width
fig, axes = plt.subplots(1, 3, figsize=(10, 4), sharex=False)

for ax, (name, data) in zip(axes, datasets.items()):
    ax.errorbar(log_alpha, data["bayes"], yerr=data["bayes_err"],
                fmt='-o', color='steelblue', label='Bayes')
    ax.errorbar(log_alpha, data["ada"], yerr=data["ada_err"],
                fmt='-s', color='orange', label='Ada')

    ax.set_title(name)
    ax.set_xlabel(r"$\log \alpha$")
    ax.grid(True, linestyle="--", alpha=0.4)

axes[0].set_ylabel("Accuracy")
axes[1].set_ylabel("")
axes[2].set_ylabel("")

# Shared legend
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, loc="lower center", ncol=2, frameon=False)

# Less horizontal breathing room  compact layout
plt.tight_layout(rect=(0, 0.10, 1, 1), w_pad=0.6)
plt.show()
```
